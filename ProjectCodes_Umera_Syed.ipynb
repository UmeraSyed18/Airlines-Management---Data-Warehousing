For Passenger table

```python
prompt = """
Generate information for 100 diverse passengers in a flight booking scenario
including PAX_ID, PAX_NAME, PAX_AGE, PAX_GENDER, PAX_PhoneNum, PAX_EmailID,
LOYALTY_POINTS. Ensure that the PAX_ID is an unique integer value and follows
a consistent format in ascending order. PAX_PhoneNum is also an integer value.
PAX_GENDER is indicated by either 'M' or 'F' (single char value). The loyalty
points are typically gained by frequent flyers and hence may be zero as well.
Ensure data diversity.
"""
_api_key = "*******************"


client = OpenAI(
    api_key=_api_key,
)

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": prompt,
        }
    ],
    model="gpt-3.5-turbo",
)

rows = chat_completion.choices[0].message.content
print(rows)
```

```python
passenger_data_variable = rows
```

```python
import csv

response_data = passenger_data_variable

passenger_data = [dict(item.strip().split(': ') for item in block.split('\n') if ':' in item) for block in response_data.split('\n\n') if block.strip()]

csv_file_path = "passenger_data.csv"

with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=passenger_data[0].keys())
    writer.writeheader()
    writer.writerows(passenger_data)

print(f"CSV file created at: {csv_file_path}")
```

For Flight table

For Cabin_Class table

```python
prompt = """
Generate 100 random data for different cabin classes in airline flights which
includes CLASS_ID, CLASS_NAME, SEAT_NUM, AVAILABLE. Ensure that CLASS_ID is an
unique integer value in ascending order resembling 1001,1002 etc. CLASS_NAME
will be either Business or Economy and SEAT_NUM would be a combination of 2
digits and 1 character. Seat numbers to be assigned considering the cabin class
(business class is usually seated 1 to 19 rows and 20 onwards it is economy
class in an airbus) Note that the availability status is indicated by 0 or 1.
Ensure data diversity.
"""
_api_key = "**************************"


client = OpenAI(
    api_key=_api_key,
)

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": prompt,
        }
    ],
    model="gpt-3.5-turbo",
)

ccrows = chat_completion.choices[0].message.content
print(ccrows)
```

```python
class_data_variable = ccrows
```

```python
class_info = class_data_variable

class_lines = class_info.strip().split('\n')


class_csv = "\n".join([",".join(part.split(': ')[1].split(',')[0].strip() for part in line.split(', ')) for line in class_lines if len(line.split(': ')) > 1])


csv_file_path = "class_data.csv"


with open(csv_file_path, 'w') as file:
    file.write("CLASS_ID, CLASS_NAME, SEAT_NUM, AVAILABLE\n" + class_csv)

print(f"CSV file created at: {csv_file_path}")
```

For Booking Table

```python
prompt = f"""
Generate 100 rows of diverse data for airline booking system. It includes the following:
BOOKING_ID: Unique integer value in ascending order.
PAX_ID: Passenger ID linked to ({passenger_data_variable})
FLIGHT_ID: Flight ID linked to ({flight_data_variable})
CLASS_ID: Class ID linked to ({class_data_variable})
SERVICE_ID: Service ID linked to ({services_data_variable})
PRICE_ID: Price ID linked to ({price_data_variable})
AIRCRAFT_ID: Aircraft ID linked to ({aircraft_data_variable})
BOOKING_DATE: Datetime value (YYYY-MM-DD hh:mm:ss) with data varying in months to predict seasonal trends
BOOKING_STATUS: Indicates "Confirmed" or "Pending" booking status
"""
_api_key = "******************************"


client = OpenAI(
    api_key=_api_key,
)

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": prompt,
        }
    ],
    model="gpt-3.5-turbo",
)

bkrows = chat_completion.choices[0].message.content
print(bkrows)
```

```python
booking_data_variable = bkrows
```

```python
import csv

booking_data = booking_data_variable

rows = [row.split('|') for row in booking_data.strip().split('\n')]


header = [col.strip() for col in rows[0]]
data = [[col.strip() for col in row] for row in rows[2:]]


csv_file_path = 'booking_data.csv'


with open(csv_file_path, 'w', newline='') as csv_file:
    writer = csv.writer(csv_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)


    writer.writerow(header)


    writer.writerows(data)

print(f"CSV file created successfully at {csv_file_path}")
```
