```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Number of Passengers in Different Age Groups

```python
data = {'age_group': ['18-30', '31-40', '41-50', '51-60', 'Above 60'],
        'number_of_passengers': [26, 18, 14, 5, 1]}

df = pd.DataFrame(data)

colors = ['orangered', 'yellowgreen', 'turquoise', 'seagreen', 'crimson']

plt.figure(figsize=(10, 6))
bars = plt.barh(df['age_group'], df['number_of_passengers'], color=colors)

for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, str(int(bar.get_width())), ha='left', va='center', color='black')

plt.title('Number of Passengers in Different Age Groups')
plt.xlabel('Number of Passengers')
plt.ylabel('Age Group')

plt.savefig('pax_agegroup.png')

plt.show()
```

Percentage of passengers with Loyalty points

```python
loyalty_data = {'With Loyalty Points': 35, 'Without Loyalty Points': 29}

fig, ax = plt.subplots()

ax.pie(loyalty_data.values(), labels=loyalty_data.keys(), autopct='%1.1f%%', colors=['gold', 'turquoise'], wedgeprops=dict(width=0.4), labeldistance=1.1)

centre_circle = plt.Circle((0, 0), 0.2, color='white', fc='white', linewidth=1.25)
ax.add_artist(centre_circle)

plt.title('Distribution of Passengers with and Without Loyalty Points (Donut Chart)')
plt.savefig('loyalty_points.png')
plt.show()
```

Short-haul, Medium-haul and Long-haul flights

```python

x_labels = ['0-3 hours', '3-6 hours', '6-12 hours']
y_values = [30, 14, 12]
custom_labels = ['Short Haul', 'Medium Haul', 'Long Haul']

plt.figure(figsize=(9, 7))

plt.bar(x_labels, y_values, color='mediumpurple')

for i, label in enumerate(custom_labels):
    plt.text(i, y_values[i] + 0.5, label, ha='center', va='bottom')

plt.xlabel('Flight Duration')
plt.ylabel('Number of Flights')
plt.title('Number of Flights Based on Duration')
plt.savefig('flight_duration.png')
plt.show()
```

Flight Routes

```python
data = [
    ('New York', 'London'), ('London', 'Paris'), ('Paris', 'Rome'),
    ('Rome', 'Madrid'), ('Madrid', 'Barcelona'), ('Barcelona', 'Amsterdam'),
    ('Amsterdam', 'Frankfurt'), ('Frankfurt', 'Dubai'), ('Dubai', 'Istanbul'),
    ('Istanbul', 'Moscow'), ('Moscow', 'Beijing'), ('Beijing', 'Tokyo'),
    ('Tokyo', 'Sydney'), ('Sydney', 'Auckland'), ('Auckland', 'Los Angeles'),
    ('Los Angeles', 'San Francisco'), ('San Francisco', 'Chicago'),
    ('Chicago', 'Toronto'), ('Toronto', 'Montreal'), ('Montreal', 'Vancouver'),
    ('Vancouver', 'Mexico City'), ('Mexico City', 'Rio de Janeiro'),
    ('Rio de Janeiro', 'Buenos Aires'), ('Buenos Aires', 'Lima'),
    ('Lima', 'Bogota'), ('Bogota', 'Santiago'), ('Santiago', 'Buenos Aires'),
    ('Buenos Aires', 'Lima'), ('Lima', 'Quito'), ('Quito', 'Bogota'),
    ('Bogota', 'Lima'), ('Lima', 'Panama City'), ('Panama City', 'Miami'),
    ('Miami', 'New York'), ('New York', 'London'), ('London', 'Paris'),
    ('Paris', 'Rome'), ('Rome', 'Madrid'), ('Madrid', 'Barcelona'),
    ('Barcelona', 'Amsterdam'), ('Amsterdam', 'Frankfurt'), ('Frankfurt', 'Dubai'),
    ('Dubai', 'Istanbul'), ('Istanbul', 'Moscow'), ('Moscow', 'Beijing'),
    ('Beijing', 'Tokyo'), ('Tokyo', 'Sydney'), ('Sydney', 'Auckland'),
    ('Auckland', 'Los Angeles'), ('Los Angeles', 'San Francisco'),
    ('San Francisco', 'Chicago'), ('Chicago', 'Toronto'), ('Toronto', 'Montreal'),
    ('Montreal', 'Vancouver'), ('Vancouver', 'Mexico City'),
    ('Mexico City', 'Rio de Janeiro')
]

df = pd.DataFrame(data, columns=['Departure', 'Arrival'])

heatmap_data = pd.crosstab(df['Departure'], df['Arrival'])

plt.figure(figsize=(12, 10))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='d', cbar_kws={'label': 'Number of Flights'})
plt.title('Flight Routes Heatmap')
plt.xlabel('Arrival Airport')
plt.ylabel('Departure Airport')
plt.savefig('flight_routes.png')
plt.show()
```

Distribution of Aircraft Types

```python
categories = ['Narrow-body jet airliner', 'Narrow-body twinjet airliner', 'Regional jet', 'Turboprop airliner',
              'Wide-body double-deck jet airliner', 'Wide-body quadjet airliner', 'Wide-body twinjet airliner']
num_aircraft = [25, 2, 9, 9, 2, 4, 14]

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(categories, num_aircraft, color=['#66b3ff', 'seagreen', 'darkorange', 'indianred', 'violet', 'deeppink', 'greenyellow'])

ax.set_xlabel('Number of Aircraft')
ax.set_title('Distribution of Aircraft Types')

for bar, label in zip(bars, num_aircraft):
    xval = bar.get_width()
    ax.text(xval + 0.5, bar.get_y() + bar.get_height()/2, str(label), ha='left', va='center')
plt.savefig('types.png')
plt.show()
```

Average vs Highest Seating Capacity of Different Aircrafts in the Fleet

```python
data = {
    'Aircraft Type': ['Narrow-body jet airliner', 'Narrow-body twinjet airliner', 'Regional jet',
                      'Turboprop airliner', 'Wide-body double-deck jet airliner', 'Wide-body quadjet airliner',
                      'Wide-body twinjet airliner'],
    'Average Seating Capacity': [150, 241, 69, 58, 714, 491, 374],
    'Max Seating Capacity': [240, 243, 104, 90, 853, 605, 440],
}

df = pd.DataFrame(data)

bar_width = 0.35
bar_positions1 = range(len(df['Aircraft Type']))
bar_positions2 = [pos + bar_width for pos in bar_positions1]

plt.figure(figsize=(12, 8))
plt.bar(bar_positions1, df['Average Seating Capacity'], width=bar_width, color='orange', label='Average Capacity')
plt.bar(bar_positions2, df['Max Seating Capacity'], width=bar_width, color='deepskyblue', label='Max Capacity')

plt.title('Average vs Max Seating Capacity Comparison')
plt.xlabel('Aircraft Type')
plt.ylabel('Seating Capacity')
plt.xticks([pos + bar_width / 2 for pos in bar_positions1], df['Aircraft Type'], rotation=45)
plt.legend()
plt.savefig('capacity.png')
plt.show()
```
